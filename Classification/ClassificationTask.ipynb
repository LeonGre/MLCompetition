```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# select the columns to use for PCA
cols = ['feature_{}'.format(i) for i in range(31)]

# create a StandardScaler object
scaler = StandardScaler()

# fit and transform the data using the StandardScaler object
data_scaled = scaler.fit_transform(features[cols])

# create a PCA object with 2 components
pca = PCA(n_components=2)

# fit and transform the data using the PCA object
data_pca = pca.fit_transform(data_scaled)

# create a scatter plot of the data
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=features["feature_30"], cmap="viridis")

# add a colorbar
plt.colorbar()

# set the axis labels
plt.xlabel("PC1")
plt.ylabel("PC2")

# annotate every single point
for i, feature in enumerate(cols):
    plt.annotate(feature, (data_pca[i, 0], data_pca[i, 1]))

# show the plot
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# select the columns to standardize
cols = ['feature_{}'.format(i) for i in range(31)]

# create a StandardScaler object
scaler = StandardScaler()

# fit and transform the data using the StandardScaler object
data_scaled = scaler.fit_transform(features[cols])

# create a new dataframe with the standardized data
features_standardized = pd.DataFrame(data_scaled, columns=cols)

# print the outer range of the distribution
low_range = features_standardized[cols].min() - (3 * features_standardized[cols].std())
high_range = features_standardized[cols].max() + (3 * features_standardized[cols].std())
print('outer range (low) of the distribution:')
print(low_range)
print('\nouter range (high) of the distribution:')
print(high_range)
```
